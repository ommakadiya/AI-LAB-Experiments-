```python
import nltk
from nltk.util import ngrams
from collections import defaultdict, Counter
import random
```

```python
nltk.download('punkt')
```

```python
corpus = """
I like Embedded System 
I like IoT 
I enjoy Edge Computing 
I like to integrate both Embedded and Iot 
I like how the Edge makes everything powerful 
Future of Tech involving all three 
"""
```

```python
tokens = nltk.word_tokenize(corpus.lower())
```

```python
# Creating the n-gram model
n = 2
bigrams = list(ngrams(tokens, n))
bigrams
```

```python
# Create the probability table
model = defaultdict(Counter)

for w1, w2 in bigrams:
    model[w1][w2] +=1

model
```

```python
def predict_nextword(word):
    word = word.lower()
    if word in model:
        next_words = model[word]
        prediction = max(next_words, key = next_words.get)
        return prediction
    else:
        return "No prediction available"
```

```python
input_word = "Computing"
print(predict_nextword(input_word))
```
