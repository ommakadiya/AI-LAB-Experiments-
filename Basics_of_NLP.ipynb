```python
#NLP : SPACY< NLTK< Gensim
```

```python
import spacy
```

```python
#load the language model
nlp = spacy.load('en_core_web_sm')
```

```python
doc=nlp(text)
```

```python
type(doc)
```

```python
doc[5]
```

```python
#Name entity Recognition: NER
for token in doc.ents:
  print(token.text, token.label_)
```

```python
#POS Tags : Part of Speech taggiing
text="is there anyway i cana go to the shopping mall at night"
doc=nlp(text)
for token in doc:
  print(token.text, "-->",token.pos_)

```

```python
#Stemming vs lemmatization
text="""I studied AI subject and then meeting Mr. Sachin in a meeting"""
#stemming
import nltk
from nltk.stem.porter import *

p_stem=PorterStemmer()
for word in text.split():
  print(word,"-->",p_stem.stem(word))

#lemmatization
for token in nlp(text):
  print(token.text,"-> ",token.pos_,
        "->",token.lemma_)
```
