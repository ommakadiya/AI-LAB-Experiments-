```python
from sklearn.feature_extraction.text import CountVectorizer

sentences = [
"My name is Om Makadiya  I am pursuing my degree in ICT ",
" Bachelor Degree is very important for job  ",
"Every Job has some criteria ",
"I am living in India for 20 years ",
"In India I am in Rajkot one of the cities in Gujarat ",
"My favourite food is Biryani ",
"One of the best biryani is from lucknow, Hyderabad and Bengal"
]

#Embeddings using TF-Bow approach
tf_vectorizer = CountVectorizer()
tf_matrix = tf_vectorizer.fit_transform(sentences)
```

```python
tf_matrix
```

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(sentences)
tfidf_similarity = cosine_similarity(tfidf_matrix)
tfidf_similarity
```

```python
#finding out the document similarity
def doc_sim(similarity_matrix,sentences):
  for i in range(len(sentences)):
    sims=similarity_matrix[i].copy()
    sims[i] = -1
    top_index = np.argmax(sims)
    print(sentences[i], "-->",sentences[top_index])
```

The similarity is based on lexical not in the meaning of the sentence
